```python
!pip install -U sentence-transformers umap-learn
from google.colab import files
import pandas as pd

uploaded = files.upload()  # Upload your cleaned reviews file
filename = next(iter(uploaded))
data = pd.read_csv(filename)

# Take first 124,000 reviews
data = data.head(124000)
reviews = data['content'].fillna('').astype(str)
import re

issue_keywords = {
    'bug': ['bug', 'bugs', 'glitch', 'glitches'],
    'error': ['error', 'errors', 'crash', 'crashes', 'fail'],
    'problem': ['problem', 'problems', 'issue', 'issues'],
    'ads': ['ad', 'ads', 'advertisement'],
    'login': ['login', 'log in', 'sign in', 'signin'],
    'payment': ['payment', 'paid', 'charge', 'billing', 'refund'],
    'performance': ['slow', 'lag', 'hang', 'freeze'],
    'update': ['update', 'version', 'patch'],
    'ui/ux': ['interface', 'design', 'layout', 'navigation', 'screen']
}
suggestion_keywords = {
        'add', 'include', 'suggest', 'should', 'please', 'feature', 'recommend',
    'option', 'would be better', 'could', 'need', 'wish', 'want',
    'request', 'provide', 'improve', 'enhance', 'make it', 'allow', 'support'

}

def tag_issue(review):
    review = review.lower()
    for label, keywords in issue_keywords.items():
        for kw in keywords:
            if re.search(r'\b' + re.escape(kw) + r'\b', review):
                return label
    return 'other'

data['issue_tag'] = reviews.apply(tag_issue)
from sentence_transformers import SentenceTransformer
import umap.umap_ as umap
import numpy as np

model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(reviews.tolist(), show_progress_bar=True)

data['embedding'] = list(embeddings)

# Append embeddings to DataFrame
embedding_array = np.array(data['embedding'].tolist())

# UMAP
umap_2d = umap.UMAP(n_neighbors=15, min_dist=0.05, metric='cosine').fit_transform(embedding_array)
data['x'] = umap_2d[:, 0]
data['y'] = umap_2d[:, 1]
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 10))
sns.scatterplot(
    data=data,
    x='x', y='y',
    hue='issue_tag',
    palette='tab10',
    s=12,
    edgecolor='none',
    alpha=0.7
)
plt.title("Clusters of User Reviews by Issue Type (124,000 Reviews)", fontsize=16)
plt.xlabel("UMAP Dimension 1")
plt.ylabel("UMAP Dimension 2")
plt.legend(title="Issue Type", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```
